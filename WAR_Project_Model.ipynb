```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score

```

```python
train_war = pd.read_csv("war_dataTrain.csv")
test_war = pd.read_csv("war_dataTest.csv")
x_train = train_war[['2012','2013','2014']]
y_train = train_war[['2015']]
x_test = test_war[['2016','2017','2018']]
y_test = test_war[['2019']]
model = linear_model.LinearRegression()
model.fit(x_train,y_train)
preds = model.predict(x_test)
#print(preds)
R_squared = r2_score(y_test,preds)
print(R_squared)
```
